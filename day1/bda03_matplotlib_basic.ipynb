##### 맷플롭립은 한글깨짐
- 나눔폰트 옵션으로 사용
- 폰트관리자 설정

````python
#Matplotlib 사용
import matplotlib.pyplot as plt #plt로 줄여씀
from matplotlib import rcParams, font_manager, rc
import warnings

#맑은 고딕으로 변경
font_path = 'C:/Windows/Fonts/malgun.ttf'
font = font_manager.FontProperties(fname=font_path, size=10).get_name()

#warning 표시끄기
warnings.simplefilter('ignore') #경고메시지 무시

#기본폰트 설정
plt.rcParams['font.family'] = font
#기본폰트 사이즈 설정
plt.rcParams['font.size'] = 14
#기본차트 사이즈 변경
plt.rcParams['figure.figsize'] = (12,5)
#기본그리드 표시
plt.rcParams['axes.grid'] = True
#한글화 후 마이너스표시 네모를 정상출력
plt.rcParams['axes.unicode_minus'] = False
````

##### 판다스로 읽은 데이터 차트화

````python
# 판다스로 데이터
import pandas as pd
````

````python
df = pd.read_csv('./data/sample_2.csv')
df
````

````python
#데이터프레임으로 맷플롭립 차트화
df[['성별','입국객수']].plot()
plt.show()
````

````python
#히스토그램
df['입국객수'].hist(bins=5)
plt.title('2019년 10월 입국객수')
plt.show()
````

````python
##스캐터차트()
x = np.random.randint(-100,100,50) #-100 ~ 100사이에 50개만 랜덤
y = np.random.randint(-200,200,50)
size = np.random.randint(0,200,50)
color = np.random.randint(-100,100,50)

plt.scatter(x, y, s=size, c=color, alpha=0.5)
plt.title('온도(화씨)')
plt.colorbar()
plt.show()
````

````python
## 파이차트
size = [2441, 2312, 1031, 1233]
label = ['A','B','AB','O']
color = ['darkmagenta','deeppink','hotpink','pink']
plt.axis('equal') #가로축, 세로축 스케일 동일
plt.pie(size, labels=label, autopct='%.1f%%', explode=(0,0,0.1,0),colors=color)
plt.legend(loc='lower left')
plt.title("혈액형별 인구율")
plt.show()
````
